```python
df.loc[df["agecategory"].str.contains("80 or older"), "agecategory"] = "80-80"
```

```python
def get_median_age(age_range):
  # Split the age range into a list of two ages
  ages = age_range.split("-")
  
  # Convert the ages to integers and take the mean
  median_age = (int(ages[0]) + int(ages[1])) / 2
  
  return median_age

# Read the .csv file into a DataFrame
# df = pd.read_csv("heart_2020_cleaned.csv")

# Create a new column called "MedianAge"
df["medianage"] = df["agecategory"].apply(get_median_age)

# Save the modified DataFrame to a new .csv file
# df.to_csv("heart_2020_cleaned_median_age.csv", index=False)
```
